```python
# https://stackoverflow.com/a/6822907
from itertools import tee

def window(iterable, size):
    iters = tee(iterable, size)
    for i in range(1, size):
        for each in iters[i:]:
            next(each, None)
    return zip(*iters)

for each in window(range(6), 3):
    print(list(each))
```

```python
from itertools import islice
# https://stackoverflow.com/a/6822773
def window(seq, n=2):
    "Returns a sliding window (of width n) over data from the iterable"
    "   s -> (s0,s1,...s[n-1]), (s1,s2,...,sn), ...                   "
    it = iter(seq)
    result = tuple(islice(it, n))
    if len(result) == n:
        yield result
    for elem in it:
        result = result[1:] + (elem,)
        yield result
```

```python
def get_sentiment(sent):
#     sent_lst = sent.split()
    for window_size in range(6, 0, -1):
        for each in window(sent.split(), window_size):
            sub_str = ' '.join(each)
            try:
                sentiment = lexicon_dict.loc[sub_str]
                sent.replace(sub_str, '')
                yield sentiment
#                 if window_size > 1:
#                     print(sub_str, ':', proprty['polarity'])
            except KeyError:
                pass
```

```python
# 'word', 'polarity', 'joy', 'fear', 'sadness', 'anger', 'surprise', 'disgust'
def gen_sentiment(text):
    text = ' '.join(text.split())
    sents = re.split(r'[,?!\.;:]', text)
    features = {
        'pos_polarity': 0,
        'neg_polarity': 0,
        'joy':0,
        'fear': 0,
        'sadness': 0,
        'anger': 0,
        'surprise': 0,
        'disgust': 0
    }
    for sent in sents:
        for sentiment in get_sentiment(sent):
            if sentiment['polarity'] == 'positive':
                features['pos_polarity'] += 1
            else:
                features['neg_polarity'] += 1
            features['joy'] += sentiment['joy']
            features['fear'] += sentiment['fear']
            features['sadness'] += sentiment['sadness']
            features['anger'] += sentiment['anger']
            features['surprise'] += sentiment['surprise']
            features['disgust'] += sentiment['disgust']
    
    return pd.Series(features)
```

```python
def cleaned_sentiment(text):
    doc = nlp(text)
    features = {
        'pos_polarity': 0,
        'neg_polarity': 0,
        'joy':0,
        'fear': 0,
        'sadness': 0,
        'anger': 0,
        'surprise': 0,
        'disgust': 0
    }
    for token in doc:
        if token.pos_ == 'ADV' or token.pos_ == 'ADJ':
            cleaned_token = token.lemma_
            try:
                sentiment = lexicon_dict.loc[cleaned_token]
                if sentiment['polarity'] == 'positive':
                    features['pos_polarity'] += 1
                else:
                    features['neg_polarity'] += 1
                features['joy'] += sentiment['joy']
                features['fear'] += sentiment['fear']
                features['sadness'] += sentiment['sadness']
                features['anger'] += sentiment['anger']
                features['surprise'] += sentiment['surprise']
                features['disgust'] += sentiment['disgust']
            except KeyError:
                pass
    return pd.Series(features)
```
